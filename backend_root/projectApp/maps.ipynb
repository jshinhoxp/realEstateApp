```python
import folium

lat = 52.3
long = 4.8

m = folium.Map(location=[lat,long],zoom_start=12)
# m.save("../templates/map.html")
```

```python
# Add markers
folium.Marker(
    location=[lat, long],
    popup="My Favorite Place"  # Optional popup text when clicking the marker
).add_to(m)

m
```

```python
# Enter the address
locations = {
   "117 Christie St, Leonia, NJ, USA",
   "82 Home Ave, Rutherford, NJ, USA"
}

import pandas as pd

df = pd.Series([locations])

from geopy.geocoders import Nominatim 
# Create geolocation
geoLoc = Nominatim(user_agent="myGeoloc")
# for each address in
for loc in locations:
   getLoc = geoLoc.geocode(str(loc))
   print(getLoc.address)
   print(f"Latitude: {getLoc.latitude} Longitude: {getLoc.longitude}")
```
